# 1. Đọc dữ liệu

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
diemTG_raw = pd.read_csv('data_TienGiang.csv', encoding = 'utf-8')
```

```python
diemTG_raw
```

# 2. Xử lý dữ liệu

### 2.1. Thêm cột số môn thi

```python
list_of_subjects = ['Toan', 'Van', 'Ngoai Ngu', 'Vat Ly', 'Hoa Hoc', 'Sinh Hoc', 'Lich Su', 'Dia Ly', 'GDCD']
diemTG_raw['So mon thi'] = diemTG_raw[list_of_subjects].notna().sum(axis=1).tolist()
diemTG_raw
```

### 2.2. Làm sạch dữ liệu

```python
diemTG = diemTG_raw.fillna(-1)
diemTG
```

# 3. Thao tác dữ liệu

### 3.1. Kiểm tra số thi sinh mỗi môn 

```python
#Đếm số lượng thí sinh thi của mỗi môn
count_toan = diemTG[diemTG['Toan'] >= 0].count()['Toan']
count_van = diemTG[diemTG['Van'] >= 0].count()['Van']
count_ngoai_ngu = diemTG[diemTG['Ngoai Ngu'] >= 0].count()['Ngoai Ngu']

count_vat_ly = diemTG[diemTG['Vat Ly'] >= 0].count()['Vat Ly']
count_hoa_hoc = diemTG[diemTG['Hoa Hoc'] >= 0].count()['Hoa Hoc']
count_sinh_hoc = diemTG[diemTG['Sinh Hoc'] >= 0].count()['Sinh Hoc']

count_lich_su = diemTG[diemTG['Lich Su'] >= 0].count()['Lich Su']
count_dia_ly = diemTG[diemTG['Dia Ly'] >= 0].count()['Dia Ly']
count_gdcd = diemTG[diemTG['GDCD'] >= 0].count()['GDCD']
```

```python
#Đếm tổng thí sinh
count_tong = diemTG.count()['SBD']
```

```python
#Tính phần trăm thí sinh thi mỗi môn
so_luong_thi = [count_toan, count_van, count_ngoai_ngu, count_vat_ly, count_hoa_hoc, 
                count_sinh_hoc, count_lich_su, count_dia_ly, count_gdcd]
phan_tram_thi = so_luong_thi / count_tong
```

#### Vẽ biểu đồ

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

ax1.bar(list_of_subjects, phan_tram_thi * 100)
ax1.set_xlabel('Môn thi')
ax1.set_ylabel('Phần trăm thi (%)')
ax1.grid(color='b', ls='-.', lw = 0.4)
ax1.set_axisbelow(True)

ax2 = ax1.twinx()
ax2.bar(list_of_subjects, so_luong_thi)
ax2.set_ylabel('Số thí sinh')
ax2.grid(color='r', ls='-', lw = 0.2)
ax2.set_axisbelow(True)

plt.title('Số thí sinh và phần trăm thí sinh thi mỗi môn')
plt.show()
```

### 3.2. Số thí sinh thi x môn

```python
# Cột cần lấy
cols = ['SBD', 'So mon thi']

# Lấy số thí sinh theo số môn thí sinh thi
diemTG_so_mon = diemTG[cols].groupby('So mon thi').count().reset_index()

# Đổi tên cột
diemTG_so_mon = diemTG_so_mon.rename(columns={'SBD': 'So thi sinh'})

diemTG_so_mon
```
